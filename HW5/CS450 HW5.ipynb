```python
import numpy as np
import numpy.linalg as la
import seaborn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def monomial(points):
    x = np.fromiter(map(lambda x: x[0], points), dtype=np.float)
    y = np.fromiter(map(lambda x: x[1], points), dtype=np.float)
    n = len(points)
    A = np.zeros((n,n))
    for i in range(n):
        A[:,i] = x**i
    coe = la.solve(A, y)
    return np.poly1d(coe[::-1])

p = monomial([(-1,1),(0,0),(1,1)])
print(p)
```

```python
def lagrange(points):
    x = np.fromiter(map(lambda x: x[0], points), dtype=np.float)
    y = np.fromiter(map(lambda x: x[1], points), dtype=np.float)
    n = len(points)
    f = np.poly1d([])
    for i in range(n):
        fi = np.poly1d([y[i]])
        for j in range(n):
            if i!=j:
                fi *= np.poly1d([1, -x[j]])/(x[i]-x[j])
        f += fi
        print(fi)
    return f
p = lagrange([(-1,1),(0,0),(1,1)])
print(p)
```
