```python
import matplotlib.pyplot as plt
import seaborn
from random import randint
%matplotlib inline
```

```python
points = [(randint(0,100), randint(0,100)) for i in range(15)]

def draw(pts, color):
    for x,y in pts:
        plt.scatter(x,y, color=color)
        plt.annotate("(%i, %i)"%(x,y),(x,y))

def undominatedPoints(pts):
    sortedPoints = sorted(pts, reverse=True)
    current_max = 0
    undominated = []
    dominated = []
    for x,y in sortedPoints:
        group = undominated if y>current_max else dominated
        group.append((x,y))
        current_max = max(y, current_max)
    return dominated, undominated

# print undominated

dominated, undominated = undominatedPoints(points)

draw(dominated, 'b')
draw(undominated, 'r')

plt.show()
```
