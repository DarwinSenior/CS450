First, we load the data in. Perturb the positions by a bit. Add up the stellar, bulge and cold mass of the galaxy. Create random training and testing datasets. Calculate the base mean squared error i.e. the expected value of the target variable from the training set is used as the prediction to evaluate the estimators. 

```python
x,y,z,n,vm,vd,rv,hmm,hmc,gmc,gvm,pvm,pmc,qvm,qmc,rvm,rmc,svm,smc,tvm,tmc,ms,mc,mb = np.loadtxt('pertm.csv', delimiter=',', skiprows=1, unpack=True) #load data
gx,gy,gz = (np.zeros(len(x)) for _ in range(3))
M = ms
#print(max(M))
for i in xrange(len(x)): #gaussian
    r = x[i]**2 + y[i]**2 + z[i]**2
    o = True
    while o:
        s = rv[i]
        rx = np.random.normal(0, s/3)
        ry = np.random.normal(0, s/3)
        rz = np.random.normal(0, s/3)
        rn = rx**2 + ry**2 + rz**2
        if(rn < s**2):
            a = rx+x[i]
            b = ry+y[i]
            c = rz+z[i]
            o = False    
    gx[i] = a
    gy[i] = b
    gz[i] = c
    
A = np.c_[x,y,z,n,vm,vd] #x,y,z,number of particles,vmax,velocity dispersion
G = np.c_[gx,gy,gz] 
B = np.c_[n,vm,vd,rv,hmm,hmc,gvm,gmc,qvm,qmc,tvm,tmc] #number of particles,vmax,velocity dispersion, virial radius, velocity,m_mean200,m_crit200,parentvmax,peakmassvmax,peakmassm_crit200
training_size = 0.45
A_train, A_test, G_train, G_test = cross_validation.train_test_split(A, G, train_size=training_size) #generate training and testing data randomly
B_train, B_test, M_train, M_test = cross_validation.train_test_split(B, M, train_size=training_size)
G_train_avg = np.mean(G_train)
M_train_avg = np.mean(M_train)
base_mse_G = mse(G_test, G_train_avg)
base_mse_M = mse(M_test, M_train_avg)
G_test_r = np.sqrt(G_test[:,0]**2 + G_test[:,1]**2 + G_test[:,2]**2) #x,y,z ---> r

print('The base MSE for the randomly chosen dataset is: x: %f, y: %f, z: %f, m: %f' % (base_mse_G[0], base_mse_G[1], base_mse_G[2], base_mse_M))
```

```python
#####helper functions#####

#gridsearch results; returns top 3 parameter combinations
def report(grid_scores, n_top=3): 
    top_scores = sorted(grid_scores, key=itemgetter(1), reverse=True)[:n_top]
    for i, score in enumerate(top_scores):
        print("Model with rank: {0}".format(i + 1))
        print("Mean validation score: {0:.3f} (std: {1:.3f})".format(
              score.mean_validation_score,
              np.std(score.cv_validation_scores)))
        print("Parameters: {0}".format(score.parameters))
        print("")
 
#generate plots
def genplots_M(M_pred, technique):
    a=np.linspace(0,max(M_test),1000)
    fig = plt.figure(1)
    ax = fig.add_subplot(111, axisbg='white')
    plt.hist2d(M_test[:], M_pred[:], bins=50, norm=LogNorm())
    plt.colorbar()
    plt.title(technique)
    plt.xlabel('$M_{true}$')
    plt.ylabel('$M_{predicted}$')
    plt.plot(a,a)
    plt.xlim([0,max(M_test)/2])
    plt.ylim([0,max(M_test)/2])
    plt.show()

def genplots_r(G_pred, technique):
    a=np.linspace(0,750,10000)
    fig = plt.figure(1)
    ax = fig.add_subplot(111, axisbg='white')
    plt.hist2d(G_test_r[:], G_pred[:], bins=40, norm=LogNorm())
    plt.colorbar()
    plt.title(technique)
    plt.xlabel('$r_{true}$')
    plt.ylabel('$r_{predicted}$')
    plt.plot(a,a,'.')
    plt.show()    
    
#calculate mse    
def mse(test, pred):
    s=0.0
    if isinstance(pred, float):
        for i in xrange(len(test)):
            s+=(test[i]-pred)**2
    else:
        for i in xrange(len(test)):
            s+=(test[i]-pred[i])**2
    return s/(len(test))

def feature_imp(importances, indices, n, method):
    plt.title(method)
    plt.bar(range(n), importances[indices],
       color="r", align="center")
    plt.xticks(range(n), indices)
    plt.xlim([-1, n])
    plt.show()
```
