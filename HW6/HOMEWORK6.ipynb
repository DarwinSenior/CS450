```python
import numpy as np
import scipy as sp
import numpy.linalg as la
import matplotlib.pyplot as plt
import seaborn
import scipy.optimize as op

%matplotlib inline
```

```python
def forwardEurler(f, y0, h, t0):
    yield(t0, y0)
    while True:
        y0 = y0+h*f(t0,y0)
        t0 += h
        yield t0,y0
        
from itertools import islice
i = 0;
t, y = zip(*list(islice(forwardEurler(lambda t,y: -5*y, 1, 0.1, 0), 40)))

x = np.linspace(0.0,4.0)
plt.plot(x, np.exp(-5*x))
plt.plot(t,y)
plt.show()

plt.semilogy(x, np.exp(-5*x))
plt.semilogy(t,y)
plt.show()
print(y[-1])
```

```python
def backwardEurler(f, y0, h, t0):
    yield(t0, y0)
    while True:
        t0 += h
        eq = lambda y1: y0+h*f(t0,y1)-y1
        y0 = op.fsolve(eq, y0)[0]
        yield t0,y0
        
from itertools import islice
i = 0;
t, y = zip(*list(islice(backwardEurler(lambda t,y: -5*y, 1, 0.1, 0), 40)))
x = np.linspace(0.0,4.0)
plt.plot(x, np.exp(-5*x))
plt.plot(t,y)
plt.show()

plt.semilogy(x, np.exp(-5*x))
plt.semilogy(t,y)
plt.show()
print(y[-1])
```

```python
def memorise(f):
    """ Memoization decorator for functions taking one or more arguments. """
    class memodict(dict):
        def __init__(self, f):
            self.f = f
        def __call__(self, *args):
            return self[args]
        def __missing__(self, key):
            ret = self[key] = self.f(*key)
            return ret
    return memodict(f)

def romberg(f, a, b, n=4):
    @memorise
    def T(i, k):
        if i==0:
            steps= (1<<k)+1
            h = (b-a)/(steps-1)
            return (sum(map(f, np.linspace(a,b, steps)))-0.5*(f(a)+f(b)))*h
        else:
            return (4**i*T(i-1, k)-T(i-1, k-1))/(4**i-1)
    return T(n, n)

from numpy.polynomial.legendre import leggauss

def guassianQuadrature(f, a, b, n=4):
    p,w = leggauss(n)
    p = (p+1)*(b-a)*0.5+a
    return np.sum(f(p)*w)*(b-a)/2


fx = lambda x: np.exp(-x)
fy = lambda y: np.exp(y)
print(romberg(fx,0,1)*romberg(fy,0,1))
print(guassianQuadrature(fx,0,1)*guassianQuadrature(fy,0,1))

fr = lambda r: r*np.exp(-r**2)
fp = lambda p: np.exp(np.sin(p)*np.cos(p))

print(romberg(fr, 0, 1)*romberg(fp,0,np.pi/2))
print(guassianQuadrature(fr, 0, 1)*guassianQuadrature(fp, 0, np.pi/2))
```

```python
def trapozoid(f, a1, b1, a2, b2, n):
    steps = (1<<n)+1
    h1 = (b1-a1)/(steps-1)
    h2 = (b2-a2)/(steps-1)
    w = np.ones(steps)
    w[0] = w[-1] = 0.5
    w2 = np.outer(w, w)
    p1 = np.outer(np.linspace(a1, b1, steps), np.ones(steps))
    p2 = np.outer(np.ones(steps), np.linspace(a2, b2, steps))
    return np.sum(np.sum(f(p1, p2)*w2))*h1*h2

def romberg2(f, a1, b1, a2, b2, n=4):
    @memorise
    def T(i, k):
        if i==0:
            return trapozoid(f, a1, b1, a2, b2, k)
        else:
            return (4**i*T(i-1, k)-T(i-1, k-1))/(4**i-1)
    return T(n, n)

from numpy.polynomial.legendre import leggauss

def GuasssianQ2d(f, a1, b1, a2, b2, n=4):
    p,w = leggauss(n)
    w2 = np.outer(w,w)
    p1 = np.outer((p+1)*(b1-a1)*0.5+a1, np.ones(n))
    p2 = np.outer(np.ones(n), (p+1)*(b2-a2)*0.5+a2)
    return np.sum(np.sum(f(p1, p2)*w2))*(b1-a1)*(b2-a2)/4

f1 = lambda x,y: np.exp(-x*y)
f2 = lambda r,p: r*np.exp(-r**2*np.sin(p)*np.cos(p))

print("Guassian Quadrature Rules")
print(GuasssianQ2d(f1, 0, 1, 0, 1))
print(GuasssianQ2d(f2, 0, 1, 0, np.pi/2))

print("Romberg + composite trapozoid")
print(romberg2(f1, 0, 1, 0, 1, n=2))
print(romberg2(f2, 0, 1, 0, np.pi/2))
```
